```python
#5
while True:
    nome = input("Digite o nome: ")
    if len(nome) > 3:
        break
    print("Nome inválido!")
while True:
    idade = int(input("Digite a idade: "))
    if idade >= 0 and idade <= 150:
        break
    print("Idade inválida!")
while True:
    salario = float(input("Digite o salário: "))
    if salario > 0:
        break
    print("Salário inválido!")
while True:
    sexo = input("Digite o sexo (f/m): ")
    if sexo == "f" or sexo == "m":
        break
    print("Sexo inválido!")
while True:
    estado = input("Digite o estado civil (s/c/v/d): ")
    if estado in ["s", "c", "v", "d"]:
        break
    print("Estado civil inválido!")
print("Dados válidos!")
```

```python
#7
number = int(input("Digite um número e eu direi o fatorial dele."))
if number < 0:
  print("Não tem fatorial de números negativos.")
else:
  fatorial = 1
  for i in range(1, number + 1):
    fatorial = fatorial * i
  print(f"O  fatorial desse número será {fatorial}.")
```

```python
#8
L = [5, 7, 2, 9, 4, 1, 3]
print("Tamanho da lista:", len(L))
maior = L[0]
for x in L:
    if x > maior:
        maior = x
menor = L[0]
for x in L:
    if x < menor:
        menor = x
print("Maior valor:", maior)
print("Menor valor:", menor)
print("Soma dos elementos:", sum(L))
print("Ordem crescente:", sorted(L))
print("Ordem decrescente:", sorted(L, reverse=True))
```
